```python
# 원형 링크드 리스트 작성
# 처음 node 생성 시 -> head에 node 저장 / node.link에 node 저장
# 마지막에 node 삽입시 -> 마지막 노드 link에 node 저장
#                           head 값은 node.link에 저장
# 중간 node 삽입시 -> 이전 노드의 link에 node 저장, node.link에 현재 노드 저장
#                           이전 노드의 link에 node 저장,
# 처음 node 삽입시 -> head에서 시작해서 마지막 노드 검색
#                            마지막 node.link에 node 저장
#                            node.link 에 head 저장, head에 node저장
```

```python
class Node:
    # seq = 0
    def __init__(self):
        self.data = {'name': None, 'score': None}
        self.link = None
```

```python
def menu_prt():
    print(' 1. 자료입력 -> ')
    print(' 2. 자료삭제 ')
    print(' 3. 자료 조회  ')  
    print(' 9. 프로그램 종료 ')  
    print('==================================')    
    print('선택 > ',end='')
    menu = input()
    return menu

def add_node():
    global HEAD, CNT
    # 생성
    node = Node() # 빈 노드 생성
    # node.data = input('자료 입력: ' ) # node.data 에 자료 입력
    node.data['name'] = input('이름 입력: ' ) # node.data 에 자료 입력
    node.data['score'] = input('점수 입력: ' ) # node.data 에 자료 입력
    CNT += 1
    # # 시퀀스 추가
    # Node.seq += 1
    # node.no = Node.seq
    # # 시퀀스 추가 종료

    # 처음 node 생성 시 -> head에 node 저장 / node.link에 node 저장
    if HEAD == None:
        HEAD = node # head에 node 저장
        node.link = node # node.link에 node 저장
        return

    elif HEAD.data['score'] > node.data['score']: # 1) HEAD 앞에 삽입하는 경우 (맨 앞에 삽입)
        last = HEAD # 마지막 node에 HEAD
        while last.link != HEAD: # 마지막 node 검색
            last = last.link
        last.link = node
        node.link = HEAD
        HEAD = node
        return
    
    curr = HEAD
    while curr.link != HEAD: #2) 중간에 삽입하는 경우 / 현재 node(curr)의 link값이 HEAD가 아니면 수행 
        if curr.data['score'] > node.data['score']: # 현재 node(curr)의 data 값이 node의 data값 보다 클 때
            node.link = curr # 삽입 node.link에 현재 node(curr) 저장
            prev.link = node # 이전 prev.link에 node 저장
            return
        prev = curr # 현재 node를 prev로 저장
        curr = curr.link # 다음 node로 이동

        
    # 3) 마지막에 노드 삽입 경우( 마지막 node.link에 node 저장 / node.link 에 head 저장)
    if curr.data['score'] > node.data['score']: # 현재 node(curr)의 data 값이 node의 data값 보다 클 때
        # 중간 삽입
        node.link = curr # 삽입 node.link에 현재 node(curr) 저장
        prev.link = node # 이전 prev.link에 node 저장
        return
    else:
        curr.link = node
        node.link = HEAD

def del_node():
    global HEAD, CNT
    name = input('삭제할 자료 입력 -> ')
    # 처음 node 삭제 -> HEAD = HEAD.link
    # 마지막 node.link 에 HEAD.link 
    if HEAD.data['name'] == name:
        CNT -= 1
        last = HEAD
        while last.link != HEAD:  # 마지막 node 검색
            last = last.link    
        if last == HEAD: # 노드가 1개 있을 때 삭제할 경우
            HEAD = None
            CNT = 0
            del last
        else:
            del last.link  # 메모리 삭제
            HEAD = last.link = HEAD.link           
        return
    
    curr = HEAD
    while curr.link != HEAD:
        # 중간 node 삭제 -> prev.link = curr.link
        if curr.data['name'] == name:
            CNT -= 1
            prev.link = curr.link
            del curr     # curr node 메모리 삭제
            return
        
        prev = curr
        curr = curr.link
        
   # 마지막 node 삭제  -> prev.link = curr.link
    if curr.data['name'] == name:
        CNT -= 1
        prev.link = curr.link
        del curr    # curr node 메모리 삭제
        return

def list_node():
    rank = 0
    curr = HEAD
    if HEAD == None:
        return
    while curr.link != HEAD:
        print(f"{CNT - rank} 등 -> {curr.data['name']} : {curr.data['score']}")
        rank += 1
        curr = curr.link
    print(f"{CNT - rank} 등 -> {curr.data['name']} : {curr.data['score']}")
```

```python
HEAD = None
CNT = 0
```

```python
while True:
    menu = menu_prt()
    if menu == '1':
        add_node()
    elif menu == '2':   
        del_node()
    elif menu == '3':
        list_node()
    elif menu == '9':
        break
```
