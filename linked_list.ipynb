```python
# 단순 링크드 리스트
# 학생들의 이름과 성적을 입력받아 저장
class Student:
    def __init__(self):
        self.name = None
        self.score = None
        self.link = None

# 성적순으로 링크드 리스트 작성
# insert_node() 수정
# 1. 학생 정보 입력
# 2. 학생 정보 삭제 (이름을 입력 받음)
# 3. 학생 정보 검색 (이름을 입력 받음)
# 4. 학생 전체 정보 출력 (등수도 출력)
# 9. 프로그램 종료
```

```python
def menu_print():
    print('=====================')
    print('학생 관리 프로그램')
    print('1. 학생 정보 입력')
    print('2. 학생 정보 삭제')
    print('3. 학생 정보 검색')
    print('4. 학생 전체 정보 출력')
    print('9. 프로그램 종료')
    print('=====================')
    print('선택 > ', end='')
    menu = input()
    return menu

def add_node():
    global head, tail
    name = input('이름 입력: ')
    score = int(input('성적 입력: '))
    node = Student()
    node.name = name
    node.score = score

    if not head: # 가장 처음일 때
        head = node
        tail = node
        return
    
    curr_node = head # 현재 노드 값 저장
    prev_node = curr_node
    while curr_node.link:
        if curr_node.score > node.score: # score 가져오는 거임, 입력할 score가 기존 노드의 score 보다 클 때
            if curr_node == head: # 처음 자료 앞에 삽입
                node.link = head
                head = node
            else:
                node.link = curr_node # 중간 노드 삽입
                prev_node.link = node 
            return

        prev_node = curr_node # 처음도 중간도 아니라면
        curr_node = curr_node.link

    # 마지막 노드의 값과 비교 후 삽입
    if curr_node.score > node.score:
        node.link = curr_node # 마지막 노드 앞 삽입
        prev_node.link = node 
    else:
        curr_node.link = node
        tail = node

def del_node(name):
    global head, tail
    node = head
    prev_node = node
    while node:
        if node.name == name: # 이름이 같으면 삭제
            # 1) 헤드를 삭제하는 경우
            if node == head:
                head = node.link
            # 2) 마지막을 삭제하는 경우
            elif node == tail:
                prev_node.link = None
                tail = prev_node
            # 3) 중간값을 삭제하는 경우
            else:
                prev_node.link = node.link
            del node
            return
            
        prev_node = node
        node = node.link

def search_node():
    pass

def list_node():
    node = head
    while node.link:
        print(f'{node.name} : {node.score}')
        node = node.link
```

```python
head = None
tail = None

while True:
    menu = menu_print()
    if menu == '1':
        add_node()
    elif menu == '2':
        name = input('삭제할 학생 이름 입력: ')
        del_node(name)
    elif menu == '3':
        search_node()
    elif menu == '4':
        list_node()
    elif menu == '9':
        break  
```
